### Load lib

```python
import pandas as pd
import numpy as np

from sklearn.metrics import f1_score
from fuzzywuzzy import fuzz
```

### Check data

```python
data = pd.read_csv('data_v2_10_oct.csv')
```

```python
data.is_duplicate.value_counts(dropna=False)
```

```python
data.dropna(inplace=True)
```

### Baseline

```python
print('PARTIAL_RATIO')
data['fuzz'] = data.apply(lambda x: fuzz.partial_ratio(x.name_1, x.name_2), axis=1)
for i in np.arange(50, 101, 2):
  print(f'{i} - {f1_score(data.is_duplicate, (data.fuzz > i).astype(int))}')
```

```python
print('TOKEN_SET_RATIO')
data['fuzz'] = data.apply(lambda x: fuzz.token_set_ratio(x.name_1, x.name_2), axis=1)
for i in np.arange(50, 101, 2):
  print(f'{i} - {f1_score(data.is_duplicate, (data.fuzz > i).astype(int))}')
```
