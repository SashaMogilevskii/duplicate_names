### Load lib

```python
import pandas as pd
import numpy as np
import pickle

from sklearn.metrics import f1_score, classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, StratifiedKFold
from catboost import CatBoostClassifier
```

### Check data

```python
data = pd.read_csv('../data/data_v2_19_oct.csv')
```

```python
data.head()
```

```python
data.name_1_upd = data.name_1_upd.apply(lambda comp_name: ''.join([alpha if alpha.isalpha() or alpha==' ' else "" for alpha in comp_name]))
data.name_2_upd = data.name_2_upd.apply(lambda comp_name: ''.join([alpha if alpha.isalpha() or alpha==' ' else "" for alpha in comp_name]))
```

```python
max_features = 550
```

```python

f1 = []

tfidf = TfidfVectorizer(max_df=0.85, min_df=10, max_features = max_features)
data_to_tfidf = pd.concat([data.name_1_upd, data.name_2_upd])
data_to_tfidf = data_to_tfidf.unique()


tfidf.fit(data_to_tfidf)
data_name_1 = pd.DataFrame(tfidf.transform(data.name_1_upd).toarray())
data_name_2 = pd.DataFrame(tfidf.transform(data.name_2_upd).toarray())

name_data = pd.concat([data_name_1, data_name_2], axis=1, ignore_index=False)
name_data.columns = [i for i in range(max_features * 2)]

model = CatBoostClassifier(task_type="GPU", random_state=1771)
model.fit(name_data, data.is_duplicate)

f1 = f1_score(data.is_duplicate, model.predict(name_data))

```

```python
pickle.dump(tfidf, open("tfidf1.pkl", "wb"))
```

```python
model.save_model("modelCatBoost.cbm")
```
